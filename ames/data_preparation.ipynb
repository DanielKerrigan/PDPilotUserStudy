```python
# give columns clearer names
df_all_og.columns = columns
```

```python
# remove space after WD in the sale type feature
df_all_og.replace({ 'Sale Type': { 'WD ': 'WD' }}, inplace=True)
```

```python
one_hot_columns = [
    'Zoning',
    'Alley',
    'Land Contour',
    'Utilities',
    'Lot Configuration',
    'Neighborhood',
    'Condition 1',
    'Condition 2',
    'Building Type',
    'House Style',
    'Roof Style',
    'Roof Material',
    'Exterior 1st',
    'Exterior 2nd',
    'Masonry Veneer',
    'Foundation',
    'Basement Finish Rating 1',
    'Basement Finish Rating 2',
    'Heating',
    'Electrical',
    'Garage Type',
    'Sale Type',
    'Sale Condition',
    'Fence',
    'Misc Feature'
]
```

```python
# one-hot encoding
df_all_one_hot = pd.get_dummies(df_all_og, columns=one_hot_columns)
```

```python
replace = {
    'Street': {
        'Grvl': 0,
        'Pave': 1
    },
    'Central Air': {
        'N': 0,
        'Y': 1
    },
    'Lot Shape': {
        'Reg': 0,
        'IR1': 1,
        'IR2': 2,
        'IR3': 3
    },
    'Land Slope': {
        'Gtl': 0,
        'Mod': 1,
        'Sev': 2
    },
    'Basement Height': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4,
        'NA': 5
    },
    'Basement Condition': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4,
        'NA': 5
    },
    'Basement Exposure': {
        'Gd': 0,
        'Av': 1,
        'Mn': 2,
        'No': 3,
        'NA': 4
    },
    'Exterior Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4
    },
    'Exterior Condition': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4
    },
    'Heating Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4
    },
    'Kitchen Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4
    },
    'Functionality': {
        'Typ': 0,
        'Min1': 1,
        'Min2': 2,
        'Mod': 3,
        'Maj1': 4,
        'Maj2': 5,
        'Sev': 6,
        'Sal': 7
    },
    'Fireplace Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4,
        'NA': 5
    },
    'Garage Finish': {
        'Fin': 0,
        'RFn': 1,
        'Unf': 2,
        'NA': 3
    },
    'Garage Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4,
        'NA': 5
    },
    'Garage Condition': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'Po': 4,
        'NA': 5
    },
    'Paved Driveway': {
        'Y': 0,
        'P': 1,
        'N': 2
    },
    'Pool Quality': {
        'Ex': 0,
        'Gd': 1,
        'TA': 2,
        'Fa': 3,
        'NA': 4
    }
}
```

```python
# ordinal encoding
df_all = df_all_one_hot.replace(to_replace=replace)
```

```python
df_all
```

```python
# handle missing values
df_all['Lot Frontage'].fillna(df_all['Lot Frontage'].median(), inplace=True)
df_all['Masonry Veneer Area'].fillna(df_all['Masonry Veneer Area'].median(), inplace=True)
df_all.loc[df_all['Garage Year'].isna(), 'Garage Year'] = df_all[df_all['Garage Year'].isna()]['Year Built']
df_all.dropna(inplace=True)
```
